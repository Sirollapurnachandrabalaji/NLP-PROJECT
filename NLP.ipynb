```python
import nltk

# Download the necessary NLTK data
nltk.download('punkt')  # For basic tokenization
nltk.download('punkt_tab')  # For sentence tokenization used by sumy

from sumy.parsers.plaintext import PlaintextParser
from sumy.nlp.tokenizers import Tokenizer
from sumy.summarizers.lsa import LsaSummarizer

def summarize_text(text, num_sentences=3):
    # Parse the text
    parser = PlaintextParser.from_string(text, Tokenizer("en"))
    summarizer = LsaSummarizer()

    # Generate summary
    summary = summarizer(parser.document, num_sentences)

    # Return summarized sentences
    return " ".join(str(sentence) for sentence in summary)

if __name__ == "__main__":
    text=input()

    print("Original Text:")
    print(text)
```

```python
print("\nSummarized Text:")
print(summarize_text(text, num_sentences=10))
```
